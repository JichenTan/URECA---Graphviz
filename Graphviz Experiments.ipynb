## Attempt 2 Neuron Network Graph

Drawing Mechanics: Each layer is a subgraph, [subgraph cluster referrence](https://www.graphviz.org/Gallery/directed/cluster.html)  

Assign splines=line to straighten the edges.

```python
#using subgraph structure to have vertical nodes and edges --> graphviz edge style?
#problematic part: line are not all the way straight

temp2="""
digraph neuron_network {
    
        
        graph [rankdir=LR splines=line]
        node [width=0.3 height=0.3 shape = circle label=""]
        edge [arrowsize = 0.5]
        
        subgraph cluster_0{
        rank = same
        node [style=invis]
        1
        2
        3
        4
        color=white
        }

        subgraph cluster_1{
        rank = same
        a1
        a2
        a3
        a4
        label="Input Layer"
        color=white
        }
        
        subgraph cluster_2{
       a1 -> {b1 b2 b3 b4 b5} 
       a2 -> {b1 b2 b3 b4 b5} 
       a3 -> {b1 b2 b3 b4 b5} 
       a4 -> {b1 b2 b3 b4 b5} 
        label = "Hidden Layer1"
        color=white
        }
        
        subgraph cluster_3{
         
        b1 -> {c1 c2 c3 c4 c5 c6 c7}
        b2 -> {c1 c2 c3 c4 c5 c6 c7}
        b3 -> {c1 c2 c3 c4 c5 c6 c7}
        b4 -> {c1 c2 c3 c4 c5 c6 c7}
        b5 -> {c1 c2 c3 c4 c5 c6 c7} 
        
        label = "Hidden Layer2"
        color=white
        }
        
        subgraph cluster_4{
        c1 -> {d1 d2 d3}   
        c2 -> {d1 d2 d3}
        c3 -> {d1 d2 d3}
        c4 -> {d1 d2 d3}   
        c5 -> {d1 d2 d3}  
        c6 -> {d1 d2 d3}
        c7 -> {d1 d2 d3}   
       
        label = "Output Layer"
        color=white
        }
        
        subgraph cluster_5{
        rank = same
        node [style=invis]
        5
        6
        7
        color=white
        }

       
        { edge [style=dotted arrowhead = normal splines="line"]  
        
       1-> a1 
       2-> a2 
       3-> a3 
       4-> a4 
       
        d1-> 5
        d2-> 6
        d3-> 7
        
       }
       
 

        
      
        
}"""

s = Source(temp2)
s
```

## Attempt 3 - Inserting picture to replace complex formula




```python
temp3="""

digraph picture {
    graph [rankdir=LR]
    node [shape=plaintext];
    
    {node []
     x1 x2 x3 y
    }
    
    {node [width=2 height=2 label=" "shape = circle image="eqn.png" imagescale=true margin=1 fixedsize=true] 
    a
    }
    
     part1 [label=<<TABLE BORDER="0" CELLBORDER="1">
      <TR><TD fixedsize="true" width="80" height="50"><IMG SRC="eqn.png"/></TD></TR>
    </TABLE>>];
    
    
    x1  -> a [label = <&omega;1>]
    x2  -> a [label = <&omega;2>]
    x3  -> a [label = <&omega;3>]
     
    a -> part1
    part1 -> y
    
    
    }"""

s = Source(temp3)
s
    
    
```
